```python
import numpy as np
```

```python
# 利用数组进行数据的处理
points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)
ys
```

```python
z = np.sqrt(xs **2 + ys ** 2)
```

```python
import matplotlib.pyplot as plt
plt.imshow(z, cmap=plt.cm.gray)
plt.show()
```

```python
# 将条件逻辑表述为数组逻辑
xarr = [1, 2, 3, 4]
yarr = [5, 6, 7, 8]
cond = [True, False, True, True]

res = [(x if c else y)
      for x, y, c in zip(xarr, yarr, cond)]
```
